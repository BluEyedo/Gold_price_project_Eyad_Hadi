# Image Segmentation using K-Means

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

## Building the model

```python
%matplotlib inline
 
# Read in the image
image = cv2.imread(R'C:\Users\FAIS\Desktop\Project AI-ML\Project-ML-image-segmentation-using-Kmeans-FaisSalaeh-ZulkifleeTelayaena\images\nature2.jpg')
 
# Change color to RGB (from BGR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# display image
plt.imshow(image)
```

```python
# Reshaping the image into a 2D array of pixels and 3 color values (RGB)
pixel_vals = image.reshape((-1,3))
 
# Convert to float type
pixel_vals = np.float32(pixel_vals)

print(pixel_vals.shape)
```

## K-Means Clustering

##### the below line of code defines the criteria for the algorithm to stop running,which will happen is 100 iterations are run or the epsilon (which is the required accuracy) becomes 85%

```python
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.85)
```

```python
from sklearn.cluster import KMeans
```

#### while KMeans(n_clusters=3)

```python
# perform k-means clustering with number of clusters and also random centres are initially choosed for k-means clustering¶
k = 3
retval, labels, centers = cv2.kmeans(pixel_vals, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
 
# convert data into 8-bit values
centers = np.uint8(centers)
segmented_data = centers[labels.flatten()]
 
# reshape data into the original image dimensions
segmented_image = segmented_data.reshape((image.shape))

plt.xlabel("k=3", fontsize=14)
plt.imshow(segmented_image)
```

#### while KMeans(n_clusters=5)

```python
k = 5
retval, labels, centers = cv2.kmeans(pixel_vals, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
 
# convert data into 8-bit values
centers = np.uint8(centers)
segmented_data = centers[labels.flatten()]
 
# reshape data into the original image dimensions
segmented_image2 = segmented_data.reshape((image.shape))

plt.xlabel("k=5", fontsize=14)
plt.imshow(segmented_image2)
```

#### while KMeans(n_clusters=8)

```python
k = 8
retval, labels, centers = cv2.kmeans(pixel_vals, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
 
# convert data into 8-bit values
centers = np.uint8(centers)
segmented_data = centers[labels.flatten()]
 
# reshape data into the original image dimensions
segmented_image3 = segmented_data.reshape((image.shape))

plt.xlabel("k=8", fontsize=14)
plt.imshow(segmented_image3)
```
